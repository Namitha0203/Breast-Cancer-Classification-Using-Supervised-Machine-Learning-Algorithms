#Classification

Dataset:The breast cancer dataset available in the sklearn library.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_breast_cancer

```

In this project, the Breast Cancer dataset from sklearn.datasets was loaded and converted into a pandas DataFrame for easier manipulation.

```python
data = load_breast_cancer()
df = pd.DataFrame(data.data, columns=data.feature_names)
df['target'] = data.target
df['target_name'] = df['target'].apply(lambda x: data.target_names[x])
df.head(10)
```

```python
df = df.drop(columns=['target_name']) #COL IS NOT REQUIRED
```

```python
#split to x and y ie, features and traget
x=df.drop('target',axis=1)
y=df['target']
```

```python
#Splitting into training and testing data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42, stratify=y)
```

#Feature Scaling using StandardScaler
To ensure all features contribute equally to the model, StandardScaler was applied to transform the data so that each feature has mean 0 and standard deviation 1.

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

#Model Building and Evaluation

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

models={
    'Logistic Regression': LogisticRegression(),
    'Decision Tree': DecisionTreeClassifier(),
    'Random Forest': RandomForestClassifier(),
    'Support Vector Machine': SVC(),
    'K-Nearest Neighbors': KNeighborsClassifier()
}

for name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    print(f"Model: {name}")
    accuracy = accuracy_score(y_test, y_pred)
    print(f"Accuracy: {accuracy}")
    print(classification_report(y_test, y_pred))
    cm = confusion_matrix(y_test, y_pred)
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                xticklabels=['Malignant', 'Benign'], yticklabels=['Malignant', 'Benign'])
    plt.title(f'Confusion Matrix - {name}')
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.show()
```

# Logistic Regression
Logistic Regression is a linear model that predicts the probability of a class using a sigmoid function. It works well when the classes are linearly separable. The breast cancer dataset is almost linearly separable, which is why Logistic Regression performs very well.

#Decision Tree Classifier
A Decision Tree splits the data into branches based on feature thresholds, forming a tree-like structure. It is easy to interpret and can capture non-linear relationships. However, Decision Trees tend to overfit, especially on small datasets like this one, which explains its lower accuracy compared to other models.

#Random Forest Classifier
Random Forest is an ensemble method that builds multiple decision trees and averages their predictions. This reduces overfitting and improves generalization. It performs well on this dataset because it captures complex patterns while maintaining stability.

#Support Vector Machine (SVM)
SVM finds the optimal hyperplane that maximizes the margin between classes. It is highly effective in high-dimensional spaces and works well when the data is clean and well-separated. Since the Breast Cancer dataset has clear class boundaries, SVM achieves very high accuracy.

#k-Nearest Neighbors (k-NN)
k-NN is a distance-based algorithm that classifies a sample based on the majority class among its nearest neighbors. It performs well when the data is scaled and when classes form distinct clusters. After scaling, k-NN performs strongly on this dataset.

#Model Comparison Summary
After training and evaluating all five models, the following observations were made:

* **Best Performing Models:** **Logistic Regression and SVM** both achieved the highest accuracy of **98.24%.** This indicates that the dataset is nearly linearly separable, making linear and margin-based models highly effective.




* **Moderate Performance:** Random Forest and KNN achieved strong accuracies of 95.61%, showing that ensemble learning and distance-based methods also work well.

* **Lowest Performance:** The **Decision Tree classifier achieved 90.35%**, the lowest among all models. This is likely due to overfitting, as single trees tend to capture noise in the data.

Overall, the dataset is well-suited for linear and margin-based classifiers, which explains the superior performance of Logistic Regression and SVM.
